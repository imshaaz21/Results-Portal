```python
import pandas as pd
import numpy as np

# Load both sheets from the Excel file
file_path = "/home/shanaaz/Downloads/Copy of Results of Focus Stride 2024.xlsx"
```

```python
physical_science_df = pd.read_excel(file_path, sheet_name="Physical Science")
biological_science_df = pd.read_excel(file_path, sheet_name="Biological Science")
```

```python
def clean_sheet(df, is_physical_science=True):
    df = df.iloc[1:].reset_index(drop=True)

    df.iloc[0, -1] = 'Rank'
    df.iloc[0, -2] = 'Z-Score'
    df.drop(columns=[df.columns[-3]], inplace=True)

    df.columns = df.iloc[0]
    df = df.iloc[1:].reset_index(drop=True)

    df.columns = df.columns.str.replace(' ', '_')

    columns_to_remove = [col for col in df.columns if col.startswith('Part') or col.startswith('Total')]
    df.drop(columns=columns_to_remove, inplace=True)

    last_columns = ['Chemistry', 'Physics', 'Combined_Maths' if is_physical_science else 'Biology', 'Z-Score', 'Rank']
    df.columns = list(df.columns[:-5]) + last_columns

    if is_physical_science:
        df['Biology'] = np.nan
    else:
        df['Combined_Maths'] = np.nan

    return df
```

```python
physical_science_cleaned = clean_sheet(physical_science_df, is_physical_science=True)
biological_science_cleaned = clean_sheet(
    biological_science_df, is_physical_science=False
)
```

```python
combined_df = pd.concat(
    [physical_science_cleaned, biological_science_cleaned], ignore_index=True
)
```

```python
desired_column_order = [
    'Index_Number', 'Name_with_Initial', 'Zone', 'Stream', 'School',
    'Combined_Maths', 'Biology', 'Chemistry', 'Physics', 'Z-Score', 'Rank'
]

combined_df = combined_df[desired_column_order]
```
